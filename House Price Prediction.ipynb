```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = pd.read_csv('Housing.csv')
```

```python
le = LabelEncoder()
data['en_mainroad'] = le.fit_transform(data['mainroad'])

le = LabelEncoder()
data['en_guestroom'] = le.fit_transform(data['guestroom'])

le = LabelEncoder()
data['en_basement'] = le.fit_transform(data['basement'])

le = LabelEncoder()
data['en_hotwaterheating'] = le.fit_transform(data['hotwaterheating'])

le = LabelEncoder()
data['en_airconditioning'] = le.fit_transform(data['airconditioning'])

le = LabelEncoder()
data['en_prefarea'] = le.fit_transform(data['prefarea'])

le = LabelEncoder()
data['en_furnishingstatus'] = le.fit_transform(data['furnishingstatus'])
```

```python
data = data.drop(['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus'], axis=1)
```

```python
data.head()
```

```python
data.dtypes
```

```python
X = data.drop('price',axis=1)
Y = data['price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
Y_predic = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(Y_test, Y_predic)
mse = mean_squared_error(Y_test, Y_predic)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, Y_predic)

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")
print(f"R² Score: {r2}")
```
